```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Customer Intelligence & Retention Dataset.csv")  
df = pd.read_excel("Customer Intelligence & Retention Dataset.excel")  
df.head(10)
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
# Remove duplicate rows from the dataset to avoid repeated records 
df = df.drop_duplicates()
```

```python
df
```

```python
# Remove rows where CustomerID is missing,
df = df.dropna(subset=['CustomerID'])  
```

```python
df
```

```python
# Convert OrderDate column to datetime format for time-based analysis 
# invalid or incorrect dates will be converted to NaT
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')

# Convert CustomerID to string to treat it as a unique identifier 
# and prevent incorrect numerical operations
df['CustomerID'] = df['CustomerID'].astype(str)
```

```python
df
```

```python
# Keep only valid transactions where Quantity and UnitPrice are greater than 0
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df
```

```python
df['revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df['year'] = df['OrderDate'].dt.year
df['month'] = df['OrderDate'].dt.month
df['year_month'] = df['OrderDate'].dt.to_period('M')
```

```python
df
```

```python
customer_df = df.groupby('CustomerID').agg({
    'revenue': 'sum',
    'OrderID': 'nunique'
}).reset_index()

customer_df.rename(columns={
    'revenue': 'total_spent',
    'OrderID': 'total_orders'
}, inplace=True)
```

```python
customer_df
```

```python
## Calculate OrderDate (days since last purchase) for each customer
latest_date = df['OrderDate'].max()

recency_df = df.groupby('CustomerID').agg({
    'OrderDate': lambda x: (latest_date - x.max()).days
}).reset_index()

recency_df.rename(columns={'OrderDate': 'recency_days'}, inplace=True)
```
